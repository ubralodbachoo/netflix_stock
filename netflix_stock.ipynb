```python
import pandas as pd
```

```python
import math
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# import datasaet from csv file, store it as pandas dataframe
data1 = pd.read_csv('Netflix.csv')
```

```python
data1.head(5)
```

```python
# make sure that Date elements are datetime objects not just strings
print('Date object type:', type(data1['Date'][0]))
```

```python
data1['Date'] = pd.to_datetime(data1['Date'])
data1 = data1.rename(columns={"Adj Close":"Adj"})
```

```python
data1.head(5)
```

```python
print('Date object type:', type(data1["Date"][3]))
```

```python
data1.columns.to_list
```

```python
# Select features (X) and target (y) - here 'Close' is the target
X = data1.drop(columns=['Close','Adj', 'Date'])
y = data1['Close']
```

```python
X.info()
```

```python
print(X.head(5))  #pandas dataframe is 2 dimentional columns with rows, columns
print(y.head(5))  #pandas series is 1 dimentional array
```

```python
#prepare datasets for modle training/testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# choose linear regression model
model1 = LinearRegression()
```

```python
# train the model
model1.fit(X_train, y_train)
```

```python
# Make predictions
predict = model1.predict(X_test)
```

```python
# seperate dataset to compare model prediction and test dataset values
compare = pd.DataFrame({'real': y_test, 'predicted' : predict})
compare['error'] = (compare['real'] - compare['predicted']).round(5)
compare
```

```python
len(compare[compare['error'] != 0.0])
```

```python
# Calculate performance metrics
mse = mean_squared_error(y_test, predict)
rmse = math.sqrt(mse)
r2 = r2_score(y_test, predict)
    
print(f"{type(model1).__name__} - Mean Squared Error: {round(rmse,4)}, R^2 Score: {round(r2,4)}")
```
